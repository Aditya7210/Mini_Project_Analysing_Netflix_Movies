```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import statistics as stat
# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
#Printing the headers of the Data
print(netflix_df.head())

#Filtering for movies
movies = netflix_df[netflix_df['type'] == 'Movie']
print(movies)

#Filter DataFrame to show movies of 1990s
movies_1990s = movies[(movies['release_year'] >= 1990) & (movies['release_year'] < 2000)]
print(movies_1990s)

#Determining the most frequest movie duration in 1990s
plt.hist(movies_1990s['duration'])
plt.xlabel('Duration')
plt.ylabel('Frequency')
plt.title('Frequency of Duration')
plt.show()
```
